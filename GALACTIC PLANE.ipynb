```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles
import pandas as pd
import matplotlib.patches as mpatches

# Read the CSV files into DataFrames
dfA = pd.read_csv("DIB-167971-final-bgt.csv", sep=';')
dfB = pd.read_csv("DIB-186745-final-bgt.csv", sep=';')

# Filter out rows where 'CW_ref' column contains 'mol'
dfA_filtered = dfA[~dfA['CW_ref'].isin(['mol'])]
dfB_filtered = dfB[~dfB['CW_ref'].isin(['mol'])]

# Extract the 'similarity' values as sets
setA = set(dfA_filtered['similarity'])
setB = set(dfB_filtered['similarity'])

# Find the intersection of the two sets
intersection = setA & setB

# Create subsets for 'n' and not 'n' values in the intersection for dfA and dfB
dfA_intersection_n = dfA_filtered[(dfA_filtered['similarity'].isin(intersection)) & (dfA_filtered['CW_ref'] == 'n')]
dfA_intersection_not_n = dfA_filtered[(dfA_filtered['similarity'].isin(intersection)) & (dfA_filtered['CW_ref'] != 'n')]

dfB_intersection_n = dfB_filtered[(dfB_filtered['similarity'].isin(intersection)) & (dfB_filtered['CW_ref'] == 'n')]
dfB_intersection_not_n = dfB_filtered[(dfB_filtered['similarity'].isin(intersection)) & (dfB_filtered['CW_ref'] != 'n')]

# Calculate the number of 'n' and non-'n' values in the intersection
n_count = len(dfA_intersection_n) + len(dfB_intersection_n)
not_n_count = len(dfA_intersection_not_n) + len(dfB_intersection_not_n)

# Calculate the number of 'n' and non-'n' values in the intersection only for dfA
n_count_A = len(dfA_intersection_n)
not_n_count_A = len(dfA_intersection_not_n)

# Calculate the number of 'n' and non-'n' values in the intersection only for dfB
n_count_B = len(dfB_intersection_n)
not_n_count_B = len(dfB_intersection_not_n)

# Calculate the counts for the unique A area excluding the intersection
dfA_only_n = dfA_filtered[(dfA_filtered['similarity'].isin(setA - intersection)) & (dfA_filtered['CW_ref'] == 'n')]
dfA_only_not_n = dfA_filtered[(dfA_filtered['similarity'].isin(setA - intersection)) & (dfA_filtered['CW_ref'] != 'n')]

n_count_only_A = len(dfA_only_n)
not_n_count_only_A = len(dfA_only_not_n)

# Define the counts for the Venn diagram
only_A = len(setA - setB)
only_B = len(setB - setA)
intersect_total = len(intersection)

# Create the Venn diagram with calculated counts
venn_diagram = venn2(subsets=(only_A, only_B, intersect_total), set_labels=(f'{len(setA)} profil DIB-HD 167971', f'{len(setB)} profil DIB-HD 186745'))
venn_circles = venn2_circles(subsets=(only_A, only_B, intersect_total))

# Manually add the labels for the intersections
venn_diagram.get_label_by_id('10').set_text('')
venn_diagram.get_label_by_id('01').set_text('')

# Get the position of the intersection text
label = venn_diagram.get_label_by_id('11')
x, y = label.get_position()
label.set_visible(False)  # Hide the original label

# Plot two larger overlapping circles to simulate the split intersection for the intersection area
plt.gca().add_patch(plt.Circle((x - 0.1, y + 0.05), 0.06, color='white', alpha=0.8))
plt.gca().add_patch(plt.Circle((x + 0.1, y - 0.05), 0.05, color='cyan', alpha=1))

# Add text labels centered within the circles for the intersection area
plt.text(x - 0.1, y + 0.05, f'{int(not_n_count/2)}', ha='center', va='center', color='black', fontsize=10)
plt.text(x + 0.1, y - 0.05, f'{int(n_count/2)}', ha='center', va='center', color='black', fontsize=10)

# Get the position of the '10' label (only A area)
label_A = venn_diagram.get_label_by_id('10')
xA, yA = label_A.get_position()

# Adjust the position for the two circles in the only A area
xA1, yA1 = xA - 0.15, yA
xA2, yA2 = xA + 0.15, yA

# Plot two larger overlapping circles to simulate the split for the only A area
plt.gca().add_patch(plt.Circle((xA1, yA1), 0.07, color='white', alpha=0.8))
plt.gca().add_patch(plt.Circle((xA2, yA2), 0.055, color='cyan', alpha=1))

# Add text labels centered within the circles for the only A area
plt.text(xA1, yA1, f'{not_n_count_only_A}', ha='center', va='center', color='black', fontsize=10)
plt.text(xA2, yA2, f'{n_count_only_A}', ha='center', va='center', color='black', fontsize=10)



# Calculate the counts for the unique B area excluding the intersection
dfB_only_n = dfB_filtered[(dfB_filtered['similarity'].isin(setB - intersection)) & (dfB_filtered['CW_ref'] == 'n')]
dfB_only_not_n = dfB_filtered[(dfB_filtered['similarity'].isin(setB - intersection)) & (dfB_filtered['CW_ref'] != 'n')]

n_count_only_B = len(dfB_only_n)
not_n_count_only_B = len(dfB_only_not_n)

# Get the position of the '01' label (only B area)
label_B = venn_diagram.get_label_by_id('01')
xB, yB = label_B.get_position()

xB1, yB1 = xB - 0.15, yB
xB2, yB2 = xB + 0.15, yB

# Plot two larger overlapping circles to simulate the split for the only B area
plt.gca().add_patch(plt.Circle((xB1 + 0.1, yB1), 0.06, color='white', alpha=0.8))
plt.gca().add_patch(plt.Circle((xB2 - 0.05, yB2), 0.055, color='cyan', alpha=1))


# Add text labels centered within the circles for the only B area
plt.text(xB1+0.1, yB1, f'{not_n_count_only_B}', ha='center', va='center', color='black', fontsize=10)
plt.text(xB2-0.05, yB2, f'{n_count_only_B}', ha='center', va='center', color='black', fontsize=10)

cyan_patch = mpatches.Patch(color='cyan', label='DIB Baru')
white_patch = mpatches.Patch(color='lightgrey', label='DIB Terkonfirmasi Katalog APO')

# Menambahkan legenda ke plot
# Set the background color to white
plt.gca().patch.set_facecolor('white')

plt.legend(handles=[white_patch, cyan_patch], loc='upper right', bbox_to_anchor=(1.3, 1))

           
# Show the plot
plt.title("Venn Diagram Katalog DIB")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load the dataset
dfA = pd.read_csv("DIB-167971.csv", sep=';')
dfA = dfA[~dfA['CW_ref'].isin(['mol'])]

# Ensure 'Tingkat Profil' column exists
if 'Tingkat Profil' in dfA.columns:
    # Sort the DataFrame by 'Tingkat Profil' column
    dfA_sorted = dfA.sort_values(by='Tingkat Profil')

    # Count the occurrences of each value in the 'Tingkat Profil' column
    tingkatan_profil_counts = dfA_sorted['Tingkat Profil'].value_counts().sort_index()

    # Define a function to format the text as 'Count (percentage)'
    def make_autopct(values):
        def my_autopct(pct):
            total = sum(values)
            val = int(round(pct * total / 100.0))
            return f'{val} ({pct:.1f}%)'
        return my_autopct

    # Define a color palette
    colors = plt.cm.Set3(range(len(tingkatan_profil_counts)))


    # Create the pie chart
    plt.figure(figsize=(7, 8))
    wedges, texts, autotexts = plt.pie(tingkatan_profil_counts, 
                                       labels=tingkatan_profil_counts.index, 
                                       autopct=make_autopct(tingkatan_profil_counts), 
                                       startangle=140, 
                                       colors=colors, 
                                       wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': 'solid', 'antialiased': True})

    # Set the title with custom font size and weight
    plt.title('Diagram Tingkatan Tingkat Kekuatan DIB-HD 167971')

    # Add custom legend with count and percentage
    handles = [Rectangle((0,0),1,1, color=color) for color in colors]
    labels = [f'{label} ({tingkatan_profil_counts[label]} profil)' for label in tingkatan_profil_counts.index]
    plt.legend(handles, labels, loc="center left", bbox_to_anchor=(0.9, 0.85), fontsize='small')

    # Display the pie chart
    plt.show()
else:
    print("The column 'Tingkat Profil' does not exist in dfA")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files into DataFrames
dfA = pd.read_csv('./DIB-167971-final-bgt.csv', sep=';')
dfB = pd.read_csv('./DIB-186745-final-bgt.csv', sep=';')

# Filter out rows where 'CW_ref' is 'mol'
dfA = dfA[~dfA['CW_ref'].isin(['mol'])]
dfB = dfB[~dfB['CW_ref'].isin(['mol'])]

# Remove rows where 'EW' is '*'
dfA = dfA[dfA['EW'] != '*']
dfB = dfB[dfB['EW'] != '*']

# Convert the 'EW' column to numeric
dfA['EW'] = pd.to_numeric(dfA['EW'], errors='coerce')
dfB['EW'] = pd.to_numeric(dfB['EW'], errors='coerce')

# Define the bins for the wavelength ranges
bins = [0, 10, 20, 30, 40, 50, 100, float('inf')]
labels = ['$W_\lambda$ < 10', '10 $\leq W_\lambda$ < 20', '20 < $\leq W_\lambda$ < 30', '30 $\leq W_\lambda$ < 40', '40 $\leq W_\lambda$ < 50', '50 $\leq W_\lambda$ < 100', '$W_\lambda\;\geq$ 100']

# Categorize the data into the bins
dfA['binned'] = pd.cut(dfA['EW'], bins=bins, labels=labels, right=False)
dfB['binned'] = pd.cut(dfB['EW'], bins=bins, labels=labels, right=False)

# Count the occurrences in each bin
countsA = dfA['binned'].value_counts().reindex(labels, fill_value=0)
countsB = dfB['binned'].value_counts().reindex(labels, fill_value=0)

# Create the figure
plt.figure(figsize=(14, 8))

# Bar width
bar_width = 0.35
index = range(len(labels))

# Create the bar charts
plt.bar(index, countsA.values, bar_width, color='turquoise', edgecolor='black', label='HD 167971')
plt.bar([i + bar_width for i in index], countsB.values, bar_width, color='lightcoral', edgecolor='black', label='HD 186745')

# Annotate each bar with its count value for dfA and dfB
for i in index:
    plt.text(i, countsA.values[i], str(countsA.values[i]), ha='center', va='bottom', fontsize=12, color='black')
    plt.text(i + bar_width, countsB.values[i], str(countsB.values[i]), ha='center', va='bottom', fontsize=12, color='black')

# Labels and title
plt.xlabel('Equivalent Width Range (mÅ)', fontsize=14)
plt.ylabel('DIB Count', fontsize=14)
plt.title('Distribusi Lebar Ekivalen Katalog DIB', fontsize=16)
plt.xticks([i + bar_width / 2 for i in index], labels, fontsize=12)

# Legend
plt.legend()

# Adjust layout for better readability
plt.tight_layout()
plt.grid()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca file CSV ke dalam DataFrame
dfA = pd.read_csv('./DIB-167971-final-bgt.csv', sep=';')
dfB = pd.read_csv('./DIB-186745-final-bgt.csv', sep=';')

# Filter out rows where 'CW_ref' is 'mol'
dfA = dfA[~dfA['CW_ref'].isin(['mol'])]
dfB = dfB[~dfB['CW_ref'].isin(['mol'])]

# Remove rows where 'EW' is '*'
dfA = dfA[dfA['EW'] != '*']
dfB = dfB[dfB['EW'] != '*']

# Convert the 'EW' and 'CW_rest' columns to numeric
dfA['EW'] = pd.to_numeric(dfA['EW'], errors='coerce')
dfB['EW'] = pd.to_numeric(dfB['EW'], errors='coerce')
dfA['CW_rest'] = pd.to_numeric(dfA['CW_rest'], errors='coerce')
dfB['CW_rest'] = pd.to_numeric(dfB['CW_rest'], errors='coerce')
dfA['CD'] = pd.to_numeric(dfA['CD'], errors='coerce')
dfB['CD'] = pd.to_numeric(dfB['CD'], errors='coerce')

# Menghitung kedalaman optik tau dari dfA['CD'] dan dfB['CD']
dfA['tau'] = dfA['CD'] / 100
dfB['tau'] = dfB['CD'] / 100

# Menghitung ekstingsi A dari tau
dfA['extinction'] = (1.086 * dfA['tau'])/1.01
dfB['extinction'] = (1.086 * dfB['tau'])/0.88

# Menghitung 1 / lambda (dalam mikrometer^-1)
dfA['inv_lambda'] = 1 / (dfA['CW_rest'] * 1e-4)  # konversi angstrom ke mikrometer
dfB['inv_lambda'] = 1 / (dfB['CW_rest'] * 1e-4)  # konversi angstrom ke mikrometer

# Membuat scatter plot
plt.figure(figsize=(14, 6))

# Plot untuk dfA
plt.subplot(1, 2, 1)
sns.scatterplot(data=dfA, x='inv_lambda', y='extinction', color='turquoise', edgecolor='black')
plt.title('Scatter Plot HD 167971')
plt.xlabel('1 / Lambda (μm^-1)')
plt.ylabel('Ekstingsi (A)')

# Plot untuk dfB
plt.subplot(1, 2, 2)
sns.scatterplot(data=dfB, x='inv_lambda', y='extinction', color='lightcoral', edgecolor='black')
plt.title('Scatter Plot HD 186745')
plt.xlabel('1 / Lambda (μm^-1)')
plt.ylabel('Ekstingsi (A)')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Membuat model regresi linier
model_A = LinearRegression()
model_B = LinearRegression()

# Melatih model regresi
model_A.fit(dfA[['extinction']], dfA['EW'])
model_B.fit(dfB[['extinction']], dfB['EW'])

# Membuat scatter plot
plt.figure(figsize=(14, 6))

# Plot untuk dfA
plt.subplot(1, 2, 1)
sns.scatterplot(data=dfA, x='extinction', y='EW', hue='EW', palette='cool', edgecolor='black')
plt.title(f'Korelasi Lebar Ekivalen dengan Ekstingsi HD 167971')
plt.xlabel('Ekstingsi MAB')
plt.ylabel('Lebar Ekivalen (mÅ)')
plt.xlim(-0.02, 0.3)
# Menambahkan garis regresi untuk dfA
plt.plot(dfA['extinction'], model_A.predict(dfA[['extinction']]), color='red', linestyle='dashed')

# Plot untuk dfB
plt.subplot(1, 2, 2)
sns.scatterplot(data=dfB, x='extinction', y='EW', hue='EW', palette='pink', edgecolor='black')
plt.title(f'Korelasi Lebar Ekivalen dengan Ekstingsi HD 186745')
plt.xlabel('Ekstingsi MAB')
plt.ylabel('Lebar Ekivalen (mÅ)')
plt.xlim(-0.02, 0.4)
# Menambahkan garis regresi untuk dfB
plt.plot(dfB['extinction'], model_B.predict(dfB[['extinction']]), color='red', linestyle='dashed')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV files into DataFrames
dfA = pd.read_csv("DIB-167971-final-bgt.csv", sep=';')
dfB = pd.read_csv("DIB-186745-final-bgt.csv", sep=';')

# Define a function to process DataFrame
def process_dataframe(df):
    # Define a function to classify rows based on the new criteria
    def classify_row(row):
        if row['CW_ref'] == 'n':
            return 'DIB Baru'
        elif pd.isna(row['Komentar']):
            return 'DIB Terkonfirmasi'
        else:
            return 'Atomik/Molekul Antarbintang'
    
    # Apply the function to each row
    df['Category'] = df.apply(classify_row, axis=1)
    
    # Calculate counts for each category
    df_counts = df['Category'].value_counts()
    return df_counts

# Process DataFrames
dfA_counts = process_dataframe(dfA)
dfB_counts = process_dataframe(dfB)

# Set the style to 'Solarize_Light2'
plt.style.use('Solarize_Light2')

# Create the figure and adjust the subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))  # Two plots side by side
fig.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05, wspace=0.3)

# Plot pie chart for dfA
wedges1, texts1, autotexts1 = ax1.pie(dfA_counts, labels=dfA_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
ax1.axis('equal')
# Add text below the plot
ax1.text(0.0, -1.4, f'{len(dfA)} Profil Absorpsi HD 167971', fontsize=12, ha='center')

# Plot pie chart for dfB
wedges2, texts2, autotexts2 = ax2.pie(dfB_counts, labels=dfB_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
ax2.axis('equal')
# Add text below the plot
ax2.text(0.0, -1.4, f'{len(dfB)} Profil Absorpsi HD 186745', fontsize=12, ha='center')

# Set the color of text to white
for autotexts in [autotexts1, autotexts2]:
    for autotext in autotexts:
        autotext.set_color('white')

# Add count values as text inside the pie chart for both plots
for autotexts, counts, wedges in zip([autotexts1, autotexts2], [dfA_counts, dfB_counts], [wedges1, wedges2]):
    for autotext, count, wedge in zip(autotexts, counts, wedges):
        if autotext.get_text().startswith('Atomik/Molekul Antarbintang'):
            # Calculate the angle of the wedge
            angle = (wedge.theta2 + wedge.theta1) / 2
            # Set the text rotation for 'Atomik/Molekul Antarbintang'
            autotext.set_rotation(angle)
            if 90 < angle < 270:
                autotext.set_rotation(angle + 180)
            autotext.set_fontstyle('italic')  # Make the text italic
            autotext.set_text(f'{count} profil')
        else:
            autotext.set_text(f'{count} profil')

# Add a single title for the entire plot
fig.suptitle('Distribusi Katalog DIB', fontsize=16)

# Set the background color to white
fig.patch.set_facecolor('white')

plt.legend()
plt.show()
print(len(dfA))
print(len(dfB))
```
